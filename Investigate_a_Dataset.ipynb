## Data Cleaning

we have problems like missing values,incorrect data types , duplicates.

## solve missing values problem

### 1- Drop some columns that we will not use or need.

```python
#for (imdb_id / tagline / homepage / overview / keywords) coulmns
df=df.drop(['imdb_id','tagline','homepage','overview','keywords'],axis=1)
```

### 2-Replace nan values for some coulmns with mode(we won't replace by mean because there are object data(categorial)).

```python
#for genres coulmn 
genres_mode=df['genres'].mode()[0]
df['genres'].fillna(genres_mode,inplace=True)
```

```python
#production_companies
production_companies_mode=df['production_companies'].mode()[0]
df['production_companies'].fillna(production_companies_mode,inplace=True)
```

### 3- Drop the null or 0 values from the data.

```python
df = df.replace(0, np.nan)
```

```python
df = df.dropna()
```

check if there are any missing values ?

```python
df.isnull().sum()
```

## Solve incorrect data type problem

the release_date came in object form which should be come in date type form.

```python
df["release_date"] = pd.to_datetime(df["release_date"])
```

```python
df.dtypes
```

## Solve duplicates problem

```python
df.duplicated().sum()
```

#### we should drop dupilcates

```python
df.drop_duplicates(inplace=True)
```

# Exploratory Data Analysis

### create method for scatterplot to only call it when we need it.

```python
def scatter(df,x,y,t,xx,yy):
    df.plot(x,y,kind='scatter',figsize=(10,10),colormap='Paired'
            ,title=t, xlabel=xx, ylabel=yy)
```

### in this phase we answer the question and discover answer by statistics or visualization.

## Q1.What is the relation between number of movies and years?

```python
df["release_year"].value_counts().plot(kind='bar',figsize=(12,10)
     ,title='The relation between number of movies and years', xlabel='year', ylabel='number of movies',colormap='Paired')
```

Coclusion: The number of movies increase over years.

## Q2.Do movies with high vote_count have high revenues?

```python
scatter(df,'vote_count','revenue','Relation between high vote_count and high revenues','Vote','Revenue')
```

Conclusion : the revenue depend on the vote counts when vote count increase then the revenue increase.

## Q3.Which year has the highest revenue?

```python
df.groupby(['release_year'])['revenue'].sum()
```

```python
df.groupby(['release_year'])['revenue'].sum().plot(kind='bar',figsize=(12,10),title='which year has heigher revenue?', 
                        xlabel='year', ylabel='revenue')
```

Conclusion: the year has the highest revenue is 2015 and revenues of moves after 2007 larger than before 2007

## Q4.Which year has the highest release of movies?

```python
df['release_year'].value_counts().plot(kind='pie',figsize=(25,15),autopct='%1.1f%%',title="Years of releasing movies ")
```

Conclusion: 2011 the year that has the maximam release of movies
